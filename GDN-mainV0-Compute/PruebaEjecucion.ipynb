```python
#prueba directa sin usar run
# definir parámetros de ejecución
gpu_n=0
DATASET=path_data+"msl"

seed=5
BATCH_SIZE=32
SLIDE_WIN=5
dim=64
out_layer_num=1
SLIDE_STRIDE=1
topk=5
out_layer_inter_dim=128
val_ratio=0.2
decay=0
EPOCH=30
report='best'

path_pattern=DATASET
COMMENT=DATASET
load_pattern=DATASET+"/pretrained/best_02.pt"
main_pattern=path+"/main.py"
```

```python
!CUDA_VISIBLE_DEVICES=$gpu_n  python $main_pattern \
        -dataset $DATASET \
        -save_path_pattern $path_pattern \
         -slide_win $SLIDE_WIN \
        -load_model_path $load_pattern
# es necesario que el modelo cargado tenga la misma configuración que la indicada al ejecutar o sino falla
#esto es lo minimo necesario, sino falla.... hay q adaptarlo para el formato de iexec de fichero de entrada y de salida.
```

```python
!CUDA_VISIBLE_DEVICES=$gpu_n  python $main_pattern \
        -dataset $DATASET \
        -save_path_pattern $path_pattern \
        -load_model_path $load_pattern \
        -slide_stride $SLIDE_STRIDE \
        -slide_win $SLIDE_WIN \
        -batch $BATCH_SIZE \
        -epoch $EPOCH \
        -comment $COMMENT \
        -random_seed $seed \
        -decay $decay \
        -dim $dim \
        -out_layer_num $out_layer_num \
        -out_layer_inter_dim $out_layer_inter_dim \
        -decay $decay \
        -val_ratio $val_ratio \
        -report $report \
        -topk $topk
```
